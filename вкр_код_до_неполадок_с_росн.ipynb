```python
# Преобразование строки, перемещая первую точку на место запятой, если она есть, и удаляя запятую
def convert_decimal(value):
    decimal_exists = False
    new_value = ""
    for char in value:
        if char == ',':
            if not decimal_exists:
                new_value += '.'
            decimal_exists = True
        elif char == '.':
            if not decimal_exists:
                new_value += ','
            else:
                new_value += char
        else:
            new_value += char
    return new_value

# Применение функции convert_decimal к каждой ячейке в датасете
portfolio = portfolio.applymap(convert_decimal)



# Вывод первых строк для проверки
print(portfolio.head())
```

```python
# Распределение потерь

# Максимальная потеря портфеля
# вероятность потери
# Значение под угрозой (VaR) – уровни достоверности: 95 %, 99 % и 99,5 %.
# предполагая, что потери нормально распределены

#derived Weights – минимальная волатильность

weights = [0.06, 0.40, 0.17, 0.13, 0.24]


pm = (portfolio * weights).pct_change().fillna(0).mean().sum().mean()

ps = np.transpose(weights) @ covariance @ weights
```

```python
pm, ps
```

```python
from scipy.stats import norm

# Compute the 95% VaR using the .ppf()
VaR_95 = norm.ppf(0.95, loc = pm, scale = ps)
# Compute the expected tail loss and the CVaR in the worst 5% of cases
tail_loss = norm.expect(lambda x: x, loc = pm, scale = ps, lb = VaR_95)
CVaR_95 = (1 / (1 - 0.95)) * tail_loss

# Plot the normal distribution histogram and add lines for the VaR and CVaR
plt.hist(norm.rvs(size = 100000, loc = pm, scale = ps), bins = 100)
plt.axvline(x = VaR_95, c='r', label = "VaR, 95% уровень уверенности")
plt.axvline(x = CVaR_95, c='b', label = "CVaR, худшие 5% результатов")
plt.legend()
plt.show()
```

```python
# Подверженность риску и потери

# Вероятность потери – вероятность негативного шока
# VaR, CVar — потери, связанные с шоком
# Толерантность к риску – Желание избежать шока.
 # Нейтральный к риску
 # нейтральный к риску
# Подверженность риску – вероятность потерь * мера потерь [VaR]
# t-распределение

from scipy.stats import t

losses = (portfolio*0.25).pct_change().sum(axis=1)
# Создать список параметров скользящего окна
mu = losses.rolling(30).mean()
sigma = losses.rolling(30).std()
rolling_parameters = [(29, mu[i], s) for i,s in enumerate(sigma)]

# Вычисляем массив VaR 99%, используя параметры скользящего окна
VaR_99 = np.array( [ t.ppf(0.99, *params)
                    for params in rolling_parameters] )

# Постройте график минимальной подверженности риску за весь период
plt.plot(losses.index, 0.01 * VaR_99 * 100000)
plt.xticks(rotation=45)
plt.show()
```

```python
#Можно сделать вывод, что подверженность риску — это просто перемасштабированная мера VaR
#максимального убытка портфеля в 99% случаев.
#Наглядно продеманстрированно насколько выше была подверженность риску
#во времена covid в 2020 году и такой же на млмент введения санкцй.
```

```python
# Fit the Student's t distribution to covid losses - 2020

crisis_losses = (portfolio.loc['1-1-2020':'31-12-2020']*0.25).pct_change().sum(axis=1)

p = t.fit(crisis_losses)

# Compute the VaR_99 for the fitted distribution
VaR_99 = t.ppf(0.99, *p)

# Use the fitted parameters and VaR_99 to compute CVaR_99
tail_loss = t.expect(lambda y: y, args = (p[0],), loc = p[1], scale = p[2], lb = VaR_99 )
CVaR_99 = (1 / (1 - 0.99)) * tail_loss
print(CVaR_99)
```

```python
# Fit the Student's t distribution 2022

crisis_losses = (portfolio.loc['1-1-2022':'31-12-2023']*0.25).pct_change().sum(axis=1)

p = t.fit(crisis_losses)

# Compute the VaR_99 for the fitted distribution
VaR_99 = t.ppf(0.99, *p)

# Use the fitted parameters and VaR_99 to compute CVaR_99
tail_loss = t.expect(lambda y: y, args = (p[0],), loc = p[1], scale = p[2], lb = VaR_99 )
CVaR_99 = (1 / (1 - 0.99)) * tail_loss
print(CVaR_99)
```

```python
!pip install statsmodels
from statsmodels.tsa.arima.model import ARIMA
```

```python
X = portfolio.SBER.values
size = int(len(X) * 0.70)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = []
for t in range(len(test)):
    model = ARIMA(history, order=(10,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)

print('predicted=%f, expected=%f' % (yhat, obs))

```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
predictions_SBER = pd.DataFrame(predictions)
predictions_SBER.index = y_test.index
predictions_SBER.columns = ['SBER']
```

```python
# Import the CovarianceShrinkage object
from pypfopt.risk_models import CovarianceShrinkage

# Create the CovarianceShrinkage instance variable
cs_actual = CovarianceShrinkage(portfolio)
cs_ARIMA = CovarianceShrinkage(ARIMA_pred_df)
cs_NN = CovarianceShrinkage(y_pred_df)

# Compute the efficient covariance matrix of returns
e_cov = cs_actual.ledoit_wolf()
e_cov_ARIMA = cs_ARIMA.ledoit_wolf()
e_cov_NN = cs_NN.ledoit_wolf()

# Display both the sample covariance_matrix and the efficient e_cov estimate
print("Efficient Covariance Matrix Actual\n", e_cov, "\n")
print("Efficient Covariance Matrix ARIMA\n", e_cov_ARIMA, "\n")
print("Efficient Covariance Matrix NN\n", e_cov_NN, "\n")
```
